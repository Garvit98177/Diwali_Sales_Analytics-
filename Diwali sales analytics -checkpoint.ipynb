```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/garvi/Desktop/project data files/Diwali Sales Data.csv",encoding='unicode_escape')
df
```

```python
df.info()
```

```python
# Remove duplicates values

df=df.drop_duplicates()
df.shape
```

```python
#check nan values

df.isna().sum()
```

```python
# drope nan values column 

df=df.drop(['Status','unnamed1'],axis=1)
df
```

```python
# Drop nan values from our data because nan values are very less 
df=df.dropna()
df
```

```python
df.isna().sum()
```

```python
# Convert Marital_Status data (0,1 into Married, Not Married)

a=[]
for i in df['Marital_Status']:
    if i==1:
        a.append("Married")
    else:
        a.append("Not Married")
df['Married/Not Married']=a
df
```

```python
# Convert Marital_Status data typ 
df['Marital_Status']=df['Marital_Status'].astype(str)
df['User_ID']=df['User_ID'].astype(str)
df.info()
```

```python
# Concatenate Gender and Age Group 

df['gen_agegr']=df['Gender']+'_'+df['Age Group']
df.head()
```

```python
# Groupby Gender and Age Group on the bases of Amount
gen_age=df.groupby(['gen_agegr'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
gen_age['Amount']=round(gen_age['Amount']/100000,2)
gen_age['Amount(in lakh)']=gen_age['Amount']
gen_age=gen_age.drop(['Amount'],axis=1)
gen_age['%']=round(gen_age['Amount(in lakh)']/sum(gen_age['Amount(in lakh)'])*100,2).astype(str)+"%"
gen_age
```

```python
# Concatenate Product_Category and Age Group
df['agegr_catag']=df['Age Group']+'_'+df['Product_Category']
df.head()
```

```python
# Groupby Product_Category and Age Group on the bases of Amount
gs=df.groupby(['agegr_catag'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
gs['Amount']=gs['Amount']/100000
gs['Amount(in lakh)']=round(gs['Amount'],2)
gs=gs.drop(['Amount'],axis=1)
gs=gs.head(20)
gs
```

```python
# Concatenate State and Product_Category
df['state_Product_Category']=df['State']+'_'+df['Product_Category']
df.head()
```

```python
# Groupby State and Product_Category on the bases of Amount
state_category=df.groupby(['state_Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
state_category['Amount']=round(state_category['Amount']/100000,2)
state_category['%']=round(state_category['Amount']/sum(state_category['Amount'])*100,2).astype(str)+"%"
gs=state_category.head(25)
gs
```
